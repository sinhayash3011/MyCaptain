```python
#importing the dependecies
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
#using pandas to read the database stored in same folder
data = pd.read_csv('mnist_test.csv')
```

```python
#viewing column heads
data.head()
```

```python
# extracting data from the dataset and viewing them up close
a = data.iloc[3,1:].values
```

```python
# reshaping the extracted data from the dataset and viewing them up close
a = a.reshape(28,28).astype('uint8')
plt.imshow(a)
```

```python
# prepare the data 
# separating labels and data values
df_x = data.iloc[:,1:]
df_y = data.iloc[:,0]
```

```python
# create test and train sizes/batches
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=4)
```

```python
#check data
x_train.head()
```

```python
#check data
y_train.head()
```

```python
# call rf classifier 
rf = RandomForestClassifier(n_estimators=100)
```

```python
# fit the model 
rf.fit(x_train, y_train)
```

```python
#prediction on test data
pred = rf.predict(x_test)
pred
```

```python
# check prediction accuracy 
s = y_test.values

#calculation number of correctly predicted values
count = 0
for i in range(len(pred)):
    if pred[i] == s[i]:
        count += 1
```
